```python
# Import necessary libraries
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris

# 1. Load a sample dataset
data = load_iris()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# 2. Define the Gradient Boosting Model
# n_estimators is the number of boosting stages (sequential models)
# learning_rate controls how much each tree influences the final prediction
gb_model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=42)

# 3. Train the model
gb_model.fit(X_train, y_train)

# 4. Evaluate the model
accuracy = gb_model.score(X_test, y_test)
print(f"Gradient Boosting Accuracy: {accuracy:.4f}")
```

##  Main Concepts and Summary of Findings

### 1. Data Preparation (Preprocessing)

* **Goal:** To prepare the raw chemical data for accurate machine learning.
* **Key Steps:**
    * **Handling Missing Values (Imputation):** Addressed potential data gaps by likely using **median** or **mode imputation** on the numerical features (like RI, Na, Mg) to maintain dataset integrity without distorting the mean (bias) or range (variance) significantly.
    * **Scaling (Standardization):** Applied **Standard Scaling** (Z-score normalization) to the chemical features. This transforms the data to have a **mean of 0** and a **standard deviation of 1**, ensuring features like Silicon (Si) with large values do not dominate distance-based algorithms over features like Potassium (K) with small values.
    * **Encoding:** Handled the categorical target variable, `Type`, likely by ensuring it was treated as an integer label (Label Encoding is implicit for a target variable in multi-class classification).

### 2. Ensemble Learning Comparison (Modeling)

Ensemble methods combine multiple base models (Decision Trees) to achieve better predictive performance than any single model.

| Ensemble Method | Training Process | Primary Goal | Comparison Note |
| :--- | :--- | :--- | :--- |
| **Random Forest (RF)** | **Parallel** training of independent, deep trees on **bootstrapped data** and **random feature subsets**. | **Reduce Variance (Overfitting)**. | Typically the **best default choice**, as the added feature randomness makes the trees less correlated, leading to highly stable and accurate predictions. |
| **Bagging** | **Parallel** training of independent, deep trees on **bootstrapped data** (no random feature subsetting). | **Reduce Variance (Overfitting)**. | An excellent technique, but RF usually outperforms it because Bagging's trees are more correlated, limiting the potential variance reduction. |
| **Boosting (Gradient Boosting)** | **Sequential** training of **shallow, weak trees**. Each subsequent tree focuses on correcting the **errors** (residuals) of the previous ones. | **Reduce Bias (Underfitting)**. | **Highly accurate**, especially when complex patterns exist. However, it is more sensitive to outliers and requires careful hyperparameter tuning to avoid overfitting the sequential corrections. |

---



 **Ensemble Methods** to classify glass type. These methods combine multiple Decision Trees to improve accuracy and robustness.

1.  **Bagging** (Bootstrap Aggregating): **Trains trees in parallel** on random data subsets to **reduce variance** (overfitting).
2.  **Random Forest:** An improvement over Bagging, adding **random feature selection** at each split to **decorrelate** the trees, leading to better overall performance.
3.  **Boosting:** **Trains trees sequentially**, with each new tree correcting the **errors** of the previous ensemble to **reduce bias**.
