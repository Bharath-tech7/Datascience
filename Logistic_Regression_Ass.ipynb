```python
# Fill missing Embarked values with the mode in Titanic_train and Titanic_test
most_embarked = Titanic_train['Embarked'].mode()[0]
Titanic_train['Embarked'].fillna(most_embarked, inplace=True)
Titanic_test['Embarked'].fillna(most_embarked, inplace=True)

print("Missing values after imputing Embarked in Titanic_train:")
print(Titanic_train.isnull().sum())
print("\nMissing values after imputing Embarked in Titanic_test:")
print(Titanic_test.isnull().sum())
```

```python
# Impute missing 'Cabin' values with a placeholder
Titanic_train['Cabin'].fillna('Unknown', inplace=True)
Titanic_test['Cabin'].fillna('Unknown', inplace=True)

print("Missing values after imputing Cabin in Titanic_train:")
print(Titanic_train.isnull().sum())
print("\nMissing values after imputing Cabin in Titanic_test:")
print(Titanic_test.isnull().sum())
```

```python
# Create dummy variables for 'Sex' and 'Embarked' in Titanic_train
train_df = pd.get_dummies(Titanic_train, columns=['Sex', 'Embarked'], drop_first=True)

# Create dummy variables for 'Sex' and 'Embarked' in Titanic_test
test_df = pd.get_dummies(Titanic_test, columns=['Sex', 'Embarked'], drop_first=True)

# Align columns between train_df and test_df - add missing columns to test_df
missing_cols = set(train_df.columns) - set(test_df.columns)
for c in missing_cols:
    if c != 'Survived': # 'Survived' is the target variable and should not be in the test set features
        test_df[c] = 0

# Align columns between train_df and test_df - remove extra columns from test_df
extra_cols = set(test_df.columns) - set(train_df.columns)
for c in extra_cols:
    test_df.drop(c, axis=1, inplace=True)

# Ensure the order of columns is the same
test_df = test_df[train_df.drop('Survived', axis=1).columns]

display(train_df.head())
display(test_df.head())
```

```python
# Create a new variable X_train by dropping the 'Survived' column from the train_df DataFrame.
X_train = train_df.drop('Survived', axis=1)

# Create a new variable y_train containing only the 'Survived' column from the train_df DataFrame.
y_train = train_df['Survived']

# Display the first few rows of X_train and y_train to verify the separation.
print("First few rows of X_train:")
display(X_train.head())
print("\nFirst few rows of y_train:")
display(y_train.head())
```

```python
# Instantiate a LogisticRegression model
model = LogisticRegression()

# Fit the model to the training data
model.fit(X_train.select_dtypes(include=np.number), y_train)
```

```python
# Make predictions on the training data
y_pred_train = model.predict(X_train.select_dtypes(include=np.number))

# Generate and print the classification report
print("Classification Report for Training Data:")
print(classification_report(y_train, y_pred_train))
```

```python
# Make predictions on the preprocessed test data using the trained model
y_pred_test = model.predict(test_df.select_dtypes(include=np.number))

# Display the first few predictions
print("First 10 predictions on the test data:")
print(y_pred_test[:10])
```

```python
from sklearn.metrics import accuracy_score
accuracy_ls = []
for thres in thresholds:
    y_pred = np.where(model.predict_proba(X_train.select_dtypes(include=np.number))[:,1]>thres,1,0)
    accuracy_ls.append(accuracy_score(y_train, y_pred, normalize=True))

accuracy_ls = pd.concat([pd.Series(thresholds), pd.Series(accuracy_ls)],
                        axis=1)
accuracy_ls.columns = ['thresholds', 'accuracy']
accuracy_ls.sort_values(by='accuracy', ascending=False, inplace=True)
accuracy_ls
```

```python
from numpy import argmax
J = tpr - fpr
ix = argmax(J)
best_thresh = thresholds[ix]
print('Best Threshold=%f' % (best_thresh))
```

```python
# Use the calculated best threshold to make predictions on the training data
threshold = best_thresh
preds = np.where(model.predict_proba(X_train.select_dtypes(include=np.number))[:,1] > threshold, 1, 0)

# Print the classification report using the training data and predictions
print(classification_report(y_train, preds))
```

```python
%%writefile app.py

import streamlit as st
import pandas as pd
import numpy as np
import joblib

# Load the trained model
model = joblib.load('logistic_regression_model.joblib')

# Define the prediction function
def predict_survival(pclass, sex_male, age, sibsp, parch, fare, embarked_Q, embarked_S):
    # Create a DataFrame with the same columns as the training data (excluding 'Survived')
    # and in the same order
    input_data = pd.DataFrame([[pclass, age, sibsp, parch, fare, sex_male, embarked_Q, embarked_S]],
                              columns=['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex_male', 'Embarked_Q', 'Embarked_S'])

    # Add other columns from the training data that are not user inputs, initialized to 0
    # This assumes that the missing columns in the input are represented as 0 in the training data after one-hot encoding,
    # or are irrelevant features dropped during training.
    # Based on the previous notebook cells, 'Name', 'Ticket', and 'Cabin' were in the training data.
    # We need to ensure the input data has the same numerical columns as the X_train used for fitting the model.
    # Let's inspect X_train's columns to be sure.
    # From the notebook state, X_train.select_dtypes(include=np.number) was used for training.
    # Let's manually list the expected numerical columns based on the model summary.
    expected_numerical_columns = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex_male', 'Embarked_Q', 'Embarked_S']

    # Ensure input_data has only the expected numerical columns in the correct order
    input_data = input_data[expected_numerical_columns]

    # Make prediction
    prediction = model.predict(input_data)
    prediction_proba = model.predict_proba(input_data)[:, 1] # Probability of survival

    return prediction[0], prediction_proba[0]


# Streamlit app layout
st.title('Titanic Survival Predictor')

st.write("""
Enter the passenger details to predict their survival on the Titanic.
""")

# Input fields
pclass = st.selectbox('Pclass', [1, 2, 3])
sex = st.selectbox('Sex', ['male', 'female'])
age = st.slider('Age', 0.42, 80.0, 30.0) # Using min and max age from describe()
sibsp = st.slider('Number of Siblings/Spouses Aboard', 0, 8, 0) # Using max SibSp from describe()
parch = st.slider('Number of Parents/Children Aboard', 0, 9, 0) # Using max Parch from describe()
fare = st.number_input('Fare', value=32.2, min_value=0.0) # Using mean Fare from describe()
embarked = st.selectbox('Port of Embarkation', ['C', 'Q', 'S'])

# Preprocess inputs
sex_male = 1 if sex == 'male' else 0
embarked_Q = 1 if embarked == 'Q' else 0
embarked_S = 1 if embarked == 'S' else 0

# Predict on button click
if st.button('Predict Survival'):
    prediction, probability = predict_survival(pclass, sex_male, age, sibsp, parch, fare, embarked_Q, embarked_S)

    st.subheader('Prediction Results:')
    if prediction == 1:
        st.success(f'The passenger is predicted to survive with a probability of {probability:.2f}.')
    else:
        st.error(f'The passenger is predicted not to survive with a probability of {1-probability:.2f}.')
```

## 1. Difference Between Precision and Recall

| Metric | Definition (What it measures) | Formula | Focus (What it minimizes) | When is it prioritized? |
| :--- | :--- | :--- | :--- | :--- |
| **Precision** | Of all instances predicted as **Positive**, how many were **actually Positive**? | $\frac{\text{True Positives}}{\text{True Positives} + \text{False Positives}}$ | **False Positives** (Type I error) | When the cost of a **False Positive** is high (e.g., spam detection: don't want to flag a legitimate email as spam). |
| **Recall** | Of all instances that are **actually Positive**, how many did the model correctly identify? | $\frac{\text{True Positives}}{\text{True Positives} + \text{False Negatives}}$ | **False Negatives** (Type II error) | When the cost of a **False Negative** is high (e.g., disease detection: don't want to miss an actual case of the disease). |

**In essence:**
* **Precision** focuses on the **quality** of the positive predictions (how trustworthy are the positive results?).
* **Recall** focuses on the **completeness** of the positive predictions (did we find all the actual positive cases?).

A high score in one often comes at the expense of the other—this is the **Precision-Recall Trade-off**.

## 2. Cross-Validation

### What is Cross-Validation?

Cross-validation is a **resampling procedure** used to evaluate machine learning models on a limited data sample. The most common method, **$k$-Fold Cross-Validation**, works as follows:
1.  The entire dataset is split into **$k$ equally-sized subsets** (or "folds").
2.  The model is trained $k$ times.
3.  In each of the $k$ iterations, one fold is reserved as the **testing set**, and the remaining $k-1$ folds are used as the **training set**.
4.  The model's performance metrics (like accuracy, precision, or recall) are recorded for each of the $k$ iterations.
5.  The final, reported performance of the model is the **average** of the $k$ recorded metric scores.

### Why is it Important in Binary Classification?

Cross-validation is crucial for:

* **Robust Performance Estimation:** It provides a **more reliable and less biased estimate** of the model's true performance on unseen data than a single train/test split. A single split might be lucky or unlucky, leading to an overly optimistic or pessimistic score.
* **Detecting Overfitting:** By repeatedly testing the model on data it hasn't been trained on, cross-validation helps to determine if the model has simply **memorized** the training data (overfitting). If the training score is consistently high but the cross-validation (test fold) scores are low, the model is likely overfit.
* **Model Selection and Hyperparameter Tuning:** It allows for a fair comparison between different models or different sets of hyperparameters for the same model. The model/hyperparameters that yield the best *average* cross-validation score are typically selected.
* **Handling Imbalanced Data (Stratified K-Fold):** In binary classification, especially when one class is much rarer than the other (**imbalanced dataset**), a variant called **Stratified $k$-Fold Cross-Validation** is used. This method ensures that **each fold maintains roughly the same proportion of the positive and negative classes** as the overall dataset, which is critical for getting a meaningful and stable performance estimate.
