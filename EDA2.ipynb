```python
from sklearn.ensemble import IsolationForest
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, LabelEncoder


# Recreate df_processed as it was not defined in this execution
# Load the dataset
df = pd.read_csv('adult_with_headers.csv')

# Apply encoding techniques (re-run from history)
categorical_cols = df.select_dtypes(include=['object']).columns
low_cardinality_cols = [col for col in categorical_cols if df[col].nunique() < 5]
high_cardinality_cols = [col for col in categorical_cols if df[col].nunique() >= 5]

if low_cardinality_cols:
    encoder_ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
    ohe_encoded_data = encoder_ohe.fit_transform(df[low_cardinality_cols])
    ohe_encoded_df = pd.DataFrame(ohe_encoded_data, columns=encoder_ohe.get_feature_names_out(low_cardinality_cols))
else:
    ohe_encoded_df = pd.DataFrame()

label_encoded_dfs = []
if high_cardinality_cols:
    for col in high_cardinality_cols:
        encoder_le = LabelEncoder()
        label_encoded_data = encoder_le.fit_transform(df[col])
        label_encoded_dfs.append(pd.DataFrame(label_encoded_data, columns=[col + '_label_encoded']))
    label_encoded_df = pd.concat(label_encoded_dfs, axis=1)
else:
    label_encoded_df = pd.DataFrame()

numerical_cols = df.select_dtypes(include=np.number).columns
df_numerical = df[numerical_cols].reset_index(drop=True)
ohe_encoded_df.reset_index(drop=True, inplace=True)
label_encoded_df.reset_index(drop=True, inplace=True)

df_processed = pd.concat([df_numerical, ohe_encoded_df, label_encoded_df], axis=1)


# 5. Outlier detection and removal
df_processed_numerical = df_processed.select_dtypes(include=['int64', 'float64'])
model_iso_forest = IsolationForest(random_state=42)
model_iso_forest.fit(df_processed_numerical)
df_processed['is_outlier'] = model_iso_forest.predict(df_processed_numerical)
df_cleaned = df_processed[df_processed['is_outlier'] == 1].copy()
```
