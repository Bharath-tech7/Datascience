```python
import pandas as pd

# Load the dataset
df = pd.read_csv('wine.csv')

# Display the first 5 rows
print("First 5 rows of the dataset:")
display(df.head())
```

```python
# Display data types and non-null values
print("\nData types and non-null values:")
display(df.info())
```

```python
# Display summary statistics
print("\nSummary statistics:")
display(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Examine the distribution of features using histograms
print("Histograms of features:")
df.hist(figsize=(15, 10))
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Display the correlation matrix
print("\nCorrelation Matrix:")
display(correlation_matrix)
```

```python
# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Wine Dataset Features')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import numpy as np

# Separate features from the target variable
feature_columns = df.drop('Type', axis=1)

# Standardize the features
scaler = StandardScaler()
scaled_df = scaler.fit_transform(feature_columns)
scaled_df = pd.DataFrame(scaled_df, columns=feature_columns.columns)

# Implement PCA
# We will start by fitting PCA to all components to determine the optimal number
pca = PCA()
pca.fit(scaled_df)

# Determine the optimal number of principal components using cumulative explained variance
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

# Visualize the cumulative explained variance
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='--')
plt.title('Cumulative Explained Variance by Number of Principal Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.grid(True)
plt.show()
```

```python

# Display the explained variance ratio for each component
print("\nExplained Variance Ratio for each Principal Component:")
display(pca.explained_variance_ratio_)
```

```python
# Display the cumulative explained variance
print("\nCumulative Explained Variance:")
display(cumulative_variance)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score

# Apply K-means clustering to the original scaled dataset
# We'll assume 3 clusters based on the 'Type' column in the original data for now,
# but in a real-world scenario, you would use techniques like the elbow method
# or silhouette analysis to determine the optimal number of clusters.
kmeans_original = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans_original.fit(scaled_df)
original_labels = kmeans_original.labels_

# Visualize the clustering results (using the first two principal components for visualization purposes)
# Although we are clustering on the original data, visualizing high-dimensional data
# is challenging. We will use the first two principal components to get a visual
# representation of the clusters in a 2D space.
pca_2_components = PCA(n_components=2)
pca_2_components_result_original = pca_2_components.fit_transform(scaled_df)

plt.figure(figsize=(10, 6))
plt.scatter(pca_2_components_result_original[:, 0], pca_2_components_result_original[:, 1], c=original_labels, cmap='viridis', marker='o', edgecolor='k', s=50)
plt.title('K-means Clustering on Original Scaled Data (Visualized with 2 PCA Components)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster Label')
plt.grid(True)
plt.show()
```

```python

# Evaluate the clustering performance
silhouette_avg_original = silhouette_score(scaled_df, original_labels)
calinski_harabasz_original = calinski_harabasz_score(scaled_df, original_labels)

print(f"\nSilhouette Score (Original Data): {silhouette_avg_original:.4f}")
print(f"Calinski-Harabasz Index (Original Data): {calinski_harabasz_original:.4f}")
```

```python
# Transform the scaled data using PCA with 3 components
pca_3_components = PCA(n_components=3)
pca_result = pca_3_components.fit_transform(scaled_df)
pca_df = pd.DataFrame(data = pca_result, columns = ['PC1', 'PC2', 'PC3'])

# Apply K-means clustering to the PCA-transformed dataset
kmeans_pca = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans_pca.fit(pca_df)
pca_labels = kmeans_pca.labels_

# Add the cluster labels to the PCA dataframe
pca_df['kmeans_label_pca'] = pca_labels

# Visualize the clustering results on PCA-transformed data (using the first two principal components)
plt.figure(figsize=(10, 6))
plt.scatter(pca_df['PC1'], pca_df['PC2'], c=pca_labels, cmap='viridis', marker='o', edgecolor='k', s=50)
plt.title('K-means Clustering on PCA-Transformed Data (First two components)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster Label')
plt.grid(True)
plt.show()
```

```python
# Evaluate the clustering performance on PCA-transformed data
silhouette_avg_pca = silhouette_score(pca_df[['PC1', 'PC2', 'PC3']], pca_labels)
calinski_harabasz_pca = calinski_harabasz_score(pca_df[['PC1', 'PC2', 'PC3']], pca_labels)

print(f"\nSilhouette Score (PCA Data): {silhouette_avg_pca:.4f}")
print(f"Calinski-Harabasz Index (PCA Data): {calinski_harabasz_pca:.4f}")
```
